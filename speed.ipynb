```python
import pandas as pd
```

```python
data = pd.read_csv("pokemon.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data["Speed"].mean()
```

```python
speed_mean = data["Speed"].mean()
```

```python
def set_speed(val):
    if val < speed_mean:
        return "Speed_low"
    elif val == speed_mean:
        return "Speed_ neutral"
    else:
        return "speed_high"
```

```python
data["Speed_high_low"] = data["Speed"].apply(set_speed)
```

```python
data.head()
```

```python
data["HP"].mean()
```

```python
HP_mean = data["HP"].mean()
```

```python
def set_HP (val):
    if val < HP_mean:
        return "HP_low"
    elif val == HP_mean:
        return "HP_neutral"
    else:
        return "HP_high"
```

```python
data["HP_high_low"] = data["HP"].apply(set_HP)
```
